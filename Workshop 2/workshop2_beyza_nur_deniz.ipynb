```python
X = []
Y = []
for x, y in points:
    X.append(x)
    Y.append(y)
X = np.array(X)
Y = np.array(Y)

X1 = np.c_[np.ones(len(X)), X]

sqr = X**2
X2 = np.c_[X1, sqr]
```
